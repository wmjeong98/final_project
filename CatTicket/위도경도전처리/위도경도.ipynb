```python
import pandas as pd
mu = pd.read_csv('./musicals.csv')
ex = pd.read_csv('./exhibitions.csv')
```

```python
import requests
import pandas as pd

def get_coords_by_place_name(query):
    url = "https://dapi.kakao.com/v2/local/search/keyword.json"
    headers = {"Authorization": "KakaoAK a62bed229b2edb385084076da980b46c"}
    params = {"query": query, "size": 1}  # size=1: 가장 관련 높은 1개만 가져오기

    response = requests.get(url, headers=headers, params=params)
    data = response.json()

    if "documents" not in data or not data["documents"]:
        print(f"⚠️ 검색 실패 또는 결과 없음: {query}")
        return None

    place = data["documents"][0]
    return {
        "lat": place.get("y"),
        "lng": place.get("x"),
        "address": place.get("address_name")
    }
```

```python
musical_lists = mu['공연장소'].tolist()
exhibition_lists = ex['공연장소'].tolist()
```

```python
results = []

for musical in musical_lists :
    place = musical
    result = get_coords_by_place_name(place)
    if result:
        results.append((place, result["lat"], result["lng"], result["address"]))
    else:
        results.append((place, "❌ 좌표 검색 실패"))
        
df = pd.DataFrame(results).reindex()

df.to_csv("musical_lat_lng.csv", index=False, encoding="utf-8-sig")
```

```python
results = []

for exhibition in exhibition_lists :
    place = exhibition
    result = get_coords_by_place_name(place)
    if result:
        results.append((place, result["lat"], result["lng"], result["address"]))
    else:
        results.append((place, "❌ 좌표 검색 실패"))
        
df = pd.DataFrame(results).reindex()

df.to_csv("exhibition_lat_lng.csv", index=False, encoding="utf-8-sig")
```
